```python
def readfile(filename):
    with open(filename,'r') as f:
        lines = []
        for line in f.readlines():
            lines.append(line.rstrip())
    return lines
```

```python
def tok(text,ts=False):

    '''
    Usage: tokenized_text = tok(text,token_list)
    If token list is not provided default one will be used instead.
    '''

    if not ts:
        ts = [',','.',';','(',')','?','!','&','%',':','*','"']

    for t in ts:
        text = text.replace(t,' ' + t + ' ')
    return text
```

```python
for k in lst:
    idx2ind = {}
    ingrs = []
    
    for i,entry in enumerate(dets):
        idx2ind[entry['id']] = i
    t = time.time()
    print ("Saving tokenized here:", 'tokenized_recipe_'+partition+str(int(k/500 + 1))+'.txt')
    f = open('tokenized_recipe_'+partition+str(int(k/500 + 1))+'.txt','w')
    
    i = k
    for i,entry in enumerate(layer1):
        '''
        if entry['id'] in dups:
            continue
        '''
        if not partition=='' and not partition == entry['partition']:
            continue
        title= entry['title']
        ingreds= entry['ingredients']
        instrs = entry['instructions']

        allingreds=''
        for ingred in ingreds:
            ingred =  ingred['text']
            allingreds+=ingred +'\n'

        allinstrs = ''
        for instr in instrs:
            instr =  instr['text']
            allinstrs+=instr + '\n'

        # find corresponding set of detected ingredients
        det_ingrs = dets[idx2ind[entry['id']]]['ingredients']
        valid = dets[idx2ind[entry['id']]]['valid']

        for j,det_ingr in enumerate(det_ingrs):
            # if detected ingredient matches ingredient text,
            # means it did not work. We skip
            if not valid[j]:
                continue
            # underscore ingredient

            det_ingr_undrs = det_ingr['text'].replace(' ','_')
            ingrs.append(det_ingr_undrs)
            allinstrs = allinstrs.replace(det_ingr['text'],det_ingr_undrs)
            
        #f.write('Title : ' + title +'\n')
        ##Change
        #print("Check",(int(k/500) + 1) )
        ##Change 
        #f.write('#')
        f.write('Ingredients:'+'\n'+allingreds)
        f.write('Instructions:'+'\n'+allinstrs+'\n')
        #if(i=1000):
        #    break;
        if (i == (k+499)):
            break;          
                      
    f.close()
    #write
    #call tok fn
    print (time.time() - t, 'seconds.')
    print ("Number of unique ingredients",len(np.unique(ingrs)))
    f = open('tokenized_recipe_'+partition+str(int(k/500 + 1))+'.txt','r')
    text = f.read()
    text = tok(text)
    f.close()
    f = open('tokenized_recipe_'+partition+str(int(k/500 + 1))+'.txt','w')
    f.write(text)
    f.close()         
```
